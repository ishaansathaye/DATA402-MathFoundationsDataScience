## Section B: Coding

```python
# 1. Implement a Function from scratch to fit a simple linear regression

import numpy as np

def estimate_coefs_slr(x, y, intercept=True):
    x = np.array(x)
    y = np.array(y)

    x_bar = np.mean(x)
    y_bar = np.mean(y)

    cov_xy = np.cov(x, y, ddof=1)[0, 1]
    var_x = np.var(x, ddof=1)

    beta_1 = cov_xy / var_x

    if intercept:
        beta_0 = y_bar - beta_1 * x_bar
    else:
        beta_0 = 0
    
    return beta_0, beta_1
```

```python
# 2. Implement a function that first a simple linear regression on data and predicts the y-vales

def predict_slr(x, y, intercept=True):
    beta_0, beta_1 = estimate_coefs_slr(x, y, intercept)
    y_hat = [beta_0 + beta_1 * i for i in x]
    return y_hat
```

```python
# 3. Write a function that fits a simple linear regression on data, predicts the y-vales and plots teh predicted values and real values.

def plot_predictions_slr(x, y, intercept=True):
    import matplotlib.pyplot as plt
    y_hat = predict_slr(x, y, intercept)
    plt.scatter(x, y, color='blue')
    plt.plot(x, y_hat, color='red')
    plt.show()
```

```python
# 4. Demonstrate the success of all 3 of your functions, with and without the intercept term, on a simple dataset.

from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])

x = df['sepal length (cm)']
y = df['sepal width (cm)']

plot_predictions_slr(x, y, intercept=True)
plot_predictions_slr(x, y, intercept=False)
```
